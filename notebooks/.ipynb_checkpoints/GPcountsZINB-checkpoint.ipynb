This notebook comapre GPcounts fit with zero-inflated negative binomial, negative binomial and Gaussian likelihoods on ScRNA-seq gene expression data in Islet $\alpha$ cell from [GSE8737 single cell RNA-seq ](https://www.ncbi.nlm.nih.gov/geo/query/acc.cgi?acc=GSE87375) dataset.

```python
import pandas as pd
import numpy as np
import gpflow 
from IPython.display import display
```

Load $\alpha$ dataset and pseudotime information

```python
Y = pd.read_csv('../data/normalized_alpha_counts.csv',index_col=[0])
X = pd.read_csv('../data/alpha_time_points.csv',index_col=[0])
```

```python
from GPcounts.GPcounts_Module import Fit_GPcounts
from helper import plot
```

```python
gene_name = ['ENSMUSG00000015879','ENSMUSG00000059173']
```

Probabilty of zeros in Fam184b and Pde1a genes

```python
Y.loc[gene_name][Y.loc[gene_name] == 0].count(axis=1)/len(Y.columns)
```

#### Fit GPcounts using full GP

```python
gp_counts = Fit_GPcounts(X,Y.loc[gene_name]) 
likelihoods = ['Zero_inflated_negative_binomial','Negative_binomial','Gaussian']
test = 'Infer_trajectory' # name of the test

for likelihood in likelihoods:
    results = gp_counts.Infer_trajectory(likelihood)
    print('The results for '+likelihood+' likelihood:')
    display(results)
    plot(gp_counts,test,likelihood,X.values,Y.loc[gene_name])
```

#### Fit GPcounts using sparse GP

```python
sparse = True
gp_counts = Fit_GPcounts(X,Y.loc[gene_name],sparse = sparse) 
likelihoods = ['Zero_inflated_negative_binomial','Negative_binomial','Gaussian']

for likelihood in likelihoods:
    sparse_results = gp_counts.Infer_trajectory(likelihood)
    print('The results for '+likelihood+' likelihood:')
    display(sparse_results)
    plot(gp_counts,test,likelihood,X.values,Y.loc[gene_name],sparse)
    
```
