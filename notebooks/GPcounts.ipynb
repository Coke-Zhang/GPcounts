GPcounts is Gaussain process regression package for count data with negative binomial and zero-inflated negative binomial likelihoods described in the paper "Gaussian process modelling of temporal and spatial counts data from RNA-seq experiments".

This notebook shows show how to run GPcounts with negative binomial likelihood and compare it with Gaussian likelihood to find differentially expressed genes using infer trajecotry, one sample and two samples cases on [fission yeast](https://bioconductor.org/packages/release/data/experiment/html/fission.html) gene expression dataset.

```python
import numpy as np
import pandas as pd
import gpflow
from IPython.display import display
```

```python
filename = '../data/fission_normalized_counts.csv'
Y = pd.read_csv(filename,index_col=[0])
X = pd.read_csv('../data/fission_col_data.csv',index_col=[0])
X = X[['minute']]
```

```python
from GPcounts.GPcounts_Module import Fit_GPcounts
```

```python
genes_name = ['SPAC11D3.01c']
gp_counts = Fit_GPcounts(X.iloc[0:18,:],Y.iloc[:,0:18].loc[genes_name]) # extract one time series
```

Infer trajectory

```python
likelihood = 'Negative_binomial' # choose the likelihood
results = gp_counts.Infer_trajectory(likelihood)
display(results)
```

plot the posterior model fit

```python
from helper import plot
```

```python
test = 'Infer_trajectory' # name of the test
plot(gp_counts,test,likelihood,X.iloc[0:18,:].values,Y.iloc[:,0:18].loc[genes_name])
```

```python
likelihood = 'Gaussian' # change the likelihood
results = gp_counts.Infer_trajectory(likelihood)
display(results)
plot(gp_counts,test,likelihood,X.iloc[0:18,:].values,Y.iloc[:,0:18].loc[genes_name])
```

One sample test

```python
likelihood = 'Negative_binomial' 
results = gp_counts.One_sample_test(likelihood)
display(results)
test = 'One_sample_test'
plot(gp_counts,test,likelihood,X.iloc[0:18,:].values,Y.iloc[:,0:18].loc[genes_name])
```

```python
likelihood = 'Gaussian' 
results = gp_counts.One_sample_test(likelihood)
display(results)
plot(gp_counts,test,likelihood,X.iloc[0:18,:].values,Y.iloc[:,0:18].loc[genes_name])
```

create a new GPcounts object to fit the two time series 

```python
gp_counts = Fit_GPcounts(X,Y.loc[genes_name])
```

Two samples test 

```python
likelihood = 'Negative_binomial' 
results = gp_counts.Two_samples_test(likelihood)
display(results)
test = 'Two_samples_test'
plot(gp_counts,test,likelihood,X.values,Y.loc[genes_name])
```

```python
likelihood = 'Gaussian' 
results = gp_counts.Two_samples_test(likelihood)
display(results)
plot(gp_counts,test,likelihood,X.values,Y.loc[genes_name])
```
